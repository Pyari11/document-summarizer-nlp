```python
import PyPDF2
from docx import Document
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from collections import defaultdict
import string

nltk.download('punkt')
nltk.download('stopwords')

file_path = input("Enter file path (.pdf, .docx, .txt): ")

text = ""

# -------- READ FILE --------

if file_path.endswith(".txt"):
    with open(file_path, "r", encoding="utf-8") as f:
        text = f.read()

elif file_path.endswith(".docx"):
    doc = Document(file_path)
    for para in doc.paragraphs:
        text += para.text + "\n"

elif file_path.endswith(".pdf"):
    reader = PyPDF2.PdfReader(file_path)
    for page in reader.pages:
        page_text = page.extract_text()
        if page_text:
            text += page_text

else:
    print("Unsupported file type")
    exit()

# -------- HEADING DETECTION --------

print("\nDetected Headings:\n")

lines = text.split("\n")

for line in lines:
    if len(line.split()) < 10 and line.isupper():
        print("▶", line)

# -------- SUMMARIZATION --------

sentences = sent_tokenize(text)

stop_words = set(stopwords.words("english"))
word_freq = defaultdict(int)

for word in word_tokenize(text.lower()):
    if word not in stop_words and word not in string.punctuation:
        word_freq[word] += 1

sentence_scores = {}

for sentence in sentences:
    for word in word_tokenize(sentence.lower()):
        if word in word_freq:
            sentence_scores[sentence] = sentence_scores.get(sentence, 0) + word_freq[word]

# automatic summary length
summary_length = max(5, len(sentences) // 4)

summary = sorted(sentence_scores, key=sentence_scores.get, reverse=True)[:summary_length]

# -------- PRINT BULLET SUMMARY --------

print("\nSUMMARY:\n")

for s in summary:
    print("•", s)
```
